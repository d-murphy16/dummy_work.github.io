- Electricity is one of the most important blessings that science has given to mankind. It has also become a part of modern life and one cannot think of a world without it.From **household chores to trade & industrial growth**, it plays a major role in our lives.

- Primary conventional energy sources for generating electricity are **Coal, Lignite, Gas, Diesel, Nuclear & Hydro** while non-conventional renewable energy sources are **Solar, Wind, Biogas etc.**

- Electricity is produced at generating plants run by companies called **GENCOS**, then is transmitted by effective transimmsion network run by companies called **TRANSCOS** & finally it is distributed to the residential/commercial premises by the **DISCOMS**

- Though **access to the electricity** has increased upto a large extent in last 20 years (**India being the energy surplus country**), its reliability, effective distribution & satisfaction to the end consumer has remain the challenge.

# **2. Problem Statement**
---

- Despite approximately 99% electrification in villages, towns, and cities across India, the challenge lies in ensuring **access to reliable electricity**.

- Frequently, issues such as prolonged **power cuts or low voltage availability/fluctuations** are encountered, particularly in semi-urban and rural areas, making demographic study imperative.

- **Demographic factors** including poverty, education/awareness levels, the efficiency of local distributors, and government policies play pivotal roles in addressing these challenges.

- Recognizing **renewable energy sources** as the potential future of distributed generation, creating awareness about these technologies is equally vital.

- Hence, this project focuses on analyzing following problem statements:

        1. Access to Reliable Electricity
        2. Availability of Major Electrical Equipments & Consumption Pattern
        3. Awareness of Alternative/Renewable Energy Sources

**<h4>Dataset: Indian Residential Energy Survey,2020 (CEEW)</h4>**

- <a href="https://www.ceew.in/">**CEEW**</a>, The Council on Energy, Environment and Water (CEEW) stands as a prominent not-for-profit policy research institution in Asia. Renowned for its commitment to fostering positive change, CEEW employs a combination of data-driven insights, integrated analysis, and strategic outreach to elucidate and transform the dynamics surrounding the utilization, recycling, and mismanagement of resources.

- As part of its extensive research endeavors, CEEW conducted the '**Indian Residential Energy Survey (IRES), 2020**' This comprehensive pan-India survey was designed to delve into the nuanced landscape of energy access and consumption patterns within Indian households. By undertaking such initiatives, CEEW contributes significantly to the discourse on sustainable energy practices and informs evidence-based policymaking in the domain of energy and the environment.
